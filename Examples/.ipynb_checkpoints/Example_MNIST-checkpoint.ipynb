# MNIST

```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
from sklearn import ensemble
import kmapper as km
from kmapper import jupyter # Creates custom CSS full-size Jupyter screen
import umap
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import aux
from sklearn import datasets
```

 This dataset is made up of 1797 8x8 images. Each image is of a hand-written digit.

```python
X, y = datasets.load_digits().data, datasets.load_digits().target
```

```python
X.shape
```

```python
#Display the first digit
plt.figure(1, figsize=(3, 3))
plt.imshow(datasets.load_digits().images[-1], cmap=plt.cm.gray_r, interpolation='nearest')
plt.show()
```

```python
y[-1]
```

### PCA

```python
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(X)
aux.visualize_scatter(principalComponents, y)
```

### t-SNE

```python
tsne = TSNE(n_components=2, perplexity=10.0)
tsne_result = tsne.fit_transform(X)
aux.visualize_scatter(tsne_result, y)
```

```python
reducer = umap.UMAP(n_neighbors = 15)
embedding = reducer.fit_transform(X)
aux.visualize_scatter(embedding, y)
```
